```python
import numpy as np
```

```python
def mrp(t, fator_desconto, recompensas_ep):
    G_t = 0.0
    for k in range(t, len(recompensas_ep)):
        G_t += fator_desconto**(k - t) * recompensas_ep[k]
    return round(G_t)
```

```python
# Exemplo de uso:
ep1 = [-5, -5, 1, 6, 0, 10, 10]
ep2 = [1,-2,1,6]
ep3 = [-5,6,0,10,-2]
print("T = 0 & GAMMA = 1")
print(mrp(0,1,ep1))
print(mrp(0,1,ep2))
print(mrp(0,1,ep3))
print("T = 1 & GAMMA = 0.8")
print(mrp(1,0.8,ep1))
print(mrp(1,0.8,ep2))
print(mrp(1,0.8,ep3))
print("T = 2 & GAMMA = 0.2")
print(mrp(2,0.2,ep1))
print(mrp(2,0.2,ep2))
print(mrp(2,0.2,ep3))
```
